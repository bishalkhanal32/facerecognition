```python
# Create a dictionary with Names (or class) as keys and list of image name of that Name (or class) as value
class_example_dict = {}

with open("/home/bishal/demoprojects/facerecognition/Datasets/Dataset.csv", "r") as f:
    for n, line in enumerate(f.readlines()):
        if n == 0:
            continue
        keyvalue = line.strip().split(",")
        if len(keyvalue) == 2:
            class_example_dict[keyvalue[1]] = class_example_dict.get(keyvalue[1], [])
            class_example_dict[keyvalue[1]].append(keyvalue[0])
        else:
            print("Debug the file: two commas in a line")

print(class_example_dict)
```

```python
# check dataset contents
for key, value in class_example_dict.items():
    print(key, len(value))

print("Class with few examples:")
for key, value in class_example_dict.items():
    if len(value) < 50:
        print(key, len(value))
 
print("Total Classes: ", len(class_example_dict.keys()))
print(class_example_dict.keys())
```

```python
from torchvision import transforms
from torchvision import datasets
import numpy as np
import pandas as pd
from facenet_pytorch import MTCNN, InceptionResnetV1
import torch
from torch.utils.data import DataLoader
import os

workers = 0 if os.name == 'nt' else 4

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)
transform = transforms.ToTensor()


def collate_fn(x):
    return x[0]

dataset_registered = datasets.ImageFolder('/home/bishal/demoprojects/facerecognition/Datasets/Faces')
dataset_registered.idx_to_class = {i:c for c, i in dataset_registered.class_to_idx.items()}
loader_registered = DataLoader(dataset_registered, collate_fn=collate_fn, num_workers=workers)

registered_embeddings = []

for x, y in loader_registered:
    x = transform(x)
    x = x.unsqueeze(dim=0)
    x = x.to(device)
    registered_embeddings.append((resnet(x).detach().cpu(), dataset_registered.idx_to_class[y]))

print(len(registered_embeddings))
# print(registered_embeddings[0].shape)
```
